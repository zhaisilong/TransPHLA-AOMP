```python
import pandas as pd
import numpy as np
from tqdm import tqdm
import os
import random
from itertools import product
from tqdm import trange
from copy import deepcopy 
from collections import Counter
import scipy.sparse as ss
from collections import OrderedDict
import json
os.chdir('/home/chujunyi/5_ZY_MHC/Anthem/')
```

```python
currentpath = os.getcwd()
unusualAA = ['B', 'J', 'O', 'U', 'X', 'Z']    
lengths = ['8', '9', '10', '11', '12', '13', '14']
pos_experimental_peptide = []

with open('%s/source/IEDBSourceprotein_Tcellassay-ligand.json' % currentpath, 'r') as f:
    dict_IDfasta = json.load(f) 
    
pwd = '/home/chujunyi/5_ZY_MHC/Anthem/Dataset/'
experimental_dataset_dict = np.load(pwd + 'experimental_dataset_dict.npy', allow_pickle = True).item()
```

```python
experimental_pos_data_pd = []
for length in lengths:
    HLAlist = list(experimental_dataset_dict[length].keys())
    for eachHLA in HLAlist:
        peptides = experimental_dataset_dict[length][eachHLA]
        for eachpeptide in peptides:
            experimental_pos_data_pd.append([eachpeptide, length, eachHLA, 1])
experimental_pos_data_pd = pd.DataFrame(experimental_pos_data_pd, columns = ['peptide', 'length', 'HLA', 'label'])
experimental_pos_data_pd.to_csv('./Dataset/experimental_pos_data.csv')
experimental_pos_data_pd
```

```python
dict_possetneg = OrderedDict()
for length in lengths:
    
    dict_possetneg[length] = OrderedDict()
    HLAlist = list(experimental_dataset_dict[length].keys())
    for eachHLA in tqdm(HLAlist):
        
        dict_possetneg[length][eachHLA] = OrderedDict()
        pos_experimental_peptide = experimental_dataset_dict[length][eachHLA]
        for eachpos in pos_experimental_peptide:
            dict_possetneg[length][eachHLA].setdefault('pos', []).append(eachpos)
        
        dict_possetneg[length][eachHLA]['neg'] = OrderedDict()    
        numpos = len(pos_experimental_peptide)
        
        
        neg_list_refine = []
        while len(neg_list_refine) <= numpos:
            thisfastacorrect = 0
            while thisfastacorrect == 0:
                choiceID = (random.sample(dict_IDfasta.keys(), 1))[0]
                choicefasta = dict_IDfasta[choiceID]
                if len(choicefasta) < int(length):
                    continue
                else:
                    havetime = 0
                    for eachAA in unusualAA:
                        if eachAA in choicefasta:
                            havetime += 1
                    if havetime >= 1:
                        continue
                startpos, stoppos = 0, int(length)
                choicelist = []
                while stoppos < len(choicefasta) + 1:
                    choicelist.append(choicefasta[startpos: stoppos])
                    startpos += 1
                    stoppos += 1
                correct = 0
                choicetime = 0
                while correct == 0:
                    choiceseg = (random.sample(choicelist, 1))[0]
                    if choiceseg not in pos_experimental_peptide and choiceseg not in neg_list_refine:
                        neg_list_refine.append(choiceseg)
                        correct += 1
                        thisfastacorrect += 1
                    else:
                        choicetime += 1
                        if choicetime == len(choicelist)*2:
                            correct += 1 
                            
        dict_possetneg[length][eachHLA]['neg'] = neg_list_refine 
```

```python
pwd = '/home/chujunyi/5_ZY_MHC/Anthem/Dataset/'
np.save(pwd + 'experimental_dict_possetneg_dict.npy', dict_possetneg)
```

```python
experimental_pos_neg_data_pd = []
for length in lengths:
    HLAlist = list(dict_possetneg[length].keys())
    for eachHLA in HLAlist:
        for type_ in ['pos', 'neg']:
            peptides = dict_possetneg[length][eachHLA][type_]
            label = [0, 1][type_ == 'pos']
            for eachpeptide in peptides:
                experimental_pos_neg_data_pd.append([eachpeptide, length, eachHLA, label])
            
experimental_pos_neg_data_pd = pd.DataFrame(experimental_pos_neg_data_pd, columns = ['peptide', 'length', 'HLA', 'label'])
experimental_pos_neg_data_pd.to_csv('./Dataset/experimental_pos_neg_data.csv')
experimental_pos_neg_data_pd
```

```python
def dict_HLAmatrix(lengths, experimental_pos_neg_dict, feature = 'AAF'):
    dict_HLAmatrix = OrderedDict()
    for length in lengths:
        dict_HLAmatrix[length] = OrderedDict()
        HLAlist = list(experimental_pos_neg_dict[length].keys())

        for eachHLA in HLAlist:
            try:
                with open('./%s/%s/%s.json' % (feature, length, eachHLA), 'r') as f:
                    fstr = f.read()
                    dict_feature_matrix = json.loads(fstr, object_pairs_hook = OrderedDict)
                dict_HLAmatrix[length][eachHLA] = dict_feature_matrix
            except:
                continue
    return dict_HLAmatrix
```

```python
def get_IC50feature(experimental_pos_neg_data_pd, dict_feature_HLAmatrix, feature, HLA_list):
    feature_list = []
    feature_name_list = []
    feature_name_list_2 = []
    feature_score_list = []
    for idx in trange(experimental_pos_neg_data_pd.shape[0]):
        item = experimental_pos_neg_data_pd.iloc[idx]
        peptide = item.peptide
        length = item.length
        feature_list_ = []
        feature_score_list_ = []
        for hla in HLA_list:
            print('-', idx, hla)
            try:
                print('---', idx, hla, type(dict_feature_HLAmatrix[length][hla]))
                try:
                    print('-----', idx, hla, type(dict_feature_HLAmatrix[length][hla]))
                    features = [eval(dict_feature_HLAmatrix[length][hla][peptide[i]][i]) for i in range(int(length))]
                except:
                    features = [dict_feature_HLAmatrix[length][hla][peptide[i]][i] for i in range(int(length))]
                
#                 features = [dict_feature_HLAmatrix[length][hla][peptide[i]][i] for i in range(int(length))]
                feature_sum, feature_mean, feature_median, feature_4quantile_pre, feature_4quantile_last, feature_min, feature_max = sum(features), np.mean(features), np.median(features), np.quantile(features, 0.25), np.quantile(features, 0.75), min(features), max(features)
                bindingscore = feature_sum/int(length)
                IC50score = math.pow(50000, (0.8 - bindingscore)/1.6)

                feature_list_.append(features)
                feature_score_list_.extend([feature_sum, feature_mean, feature_median, feature_4quantile_pre, feature_4quantile_last, feature_min, feature_max, IC50score])
                
                if idx == 0:
                    names = [feature + '_' + hla + '_sum', feature + '_' + hla + '_mean', feature + '_' + hla + '_median', feature + '_' + hla + '_4quantile_pre', feature + '_' + hla + '_4quantile_last', feature + '_' + hla + '_min', feature + '_' + hla + '_max', feature + '_' + hla + '_IC50score']
                    feature_name_list.extend(names)
                    
                    feature_name_list_2.append(feature + '_' + hla)
#                 print(hla)
            except:
                continue
        feature_list.append(feature_list_)
        feature_score_list.append(feature_score_list_)
    print(feature_name_list, pd.DataFrame(feature_score_list).shape)
    feature_score_pd = pd.DataFrame(feature_score_list, columns = feature_name_list)
    feature_pd = pd.DataFrame(feature_list, columns = feature_name_list_2)
    feature_score_pd = feature_score_pd.set_index(data.peptide)
    feature_pd = feature_pd.set_index(data.peptide)
    return feature_score_pd, feature_pd
```

```python
def get_feature(experimental_pos_neg_data_pd, dict_feature_HLAmatrix, feature, HLA_list):
    feature_list = []
    feature_name_list = []
    feature_name_list_2 = []
    feature_score_list = []
    for idx in trange(experimental_pos_neg_data_pd.shape[0]):
        item = experimental_pos_neg_data_pd.iloc[idx]
        peptide = item.peptide
        length = item.length
        feature_list_ = []
        feature_score_list_ = []
        for hla in HLA_list:
            try:
                
                try:
                    features = [eval(dict_feature_HLAmatrix[length][hla][peptide[i]][i]) for i in range(int(length))]
                except:
                    features = [dict_feature_HLAmatrix[length][hla][peptide[i]][i] for i in range(int(length))]

                feature_sum, feature_mean, feature_median, feature_4quantile_pre, feature_4quantile_last, feature_min, feature_max = sum(features), np.mean(features), np.median(features), np.quantile(features, 0.25), np.quantile(features, 0.75), min(features), max(features)
                
                feature_list_.append(features)
                feature_score_list_.extend([feature_sum, feature_mean, feature_median, feature_4quantile_pre, feature_4quantile_last, feature_min, feature_max])
                
                if idx == 0:
                    names = [feature + '_' + hla + '_sum', feature + '_' + hla + '_mean', feature + '_' + hla + '_median', feature + '_' + hla + '_4quantile_pre', feature + '_' + hla + '_4quantile_last', feature + '_' + hla + '_min', feature + '_' + hla + '_max']
                    feature_name_list.extend(names)
                    
                    feature_name_list_2.append(feature + '_' + hla)
            except:
                continue
        feature_list.append(feature_list_)
        feature_score_list.append(feature_score_list_)
        
    feature_score_pd = pd.DataFrame(feature_score_list, columns = feature_name_list)
    feature_pd = pd.DataFrame(feature_list, columns = feature_name_list_2)
    feature_score_pd = feature_score_pd.set_index(data.peptide)
    feature_pd = feature_pd.set_index(data.peptide)
    return feature_score_pd, feature_pd
```
